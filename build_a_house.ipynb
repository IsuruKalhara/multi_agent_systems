```python
from mesa import Agent, Model
from mesa.time import RandomActivation
import random

class Nuwan(Agent):
    """Our role model with fixed initial wealth."""
    def __init__(self, unique_id, model):
        super().__init__(unique_id, model)
        
        # Life goals
        self.bought_a_land = False
        self.build_a_house = False
        self.house_carpentering = False
        self.house_painting = False
        self.bought_a_car = False
        self.completed = False
        
        # in process
        self.building_the_house = False
        self.carpentering_the_house = False
        self.painting_the_house = False
        
        
        self.wealth = random.randint(5000000,7000000)
        self.salary = 50000
        self.expenses = 0
        
        print("Nuwan has ", self.wealth, "much of money")
        print("He needs to buy a land, build a house and buy a car")
        print("He got a job with a monthly salary of ",self.salary)
        print("He has no monthly expenses yet")
        print("-------------------------------------------------------\n")
        print("He found a land. But the land price increases every month")
        print("-------------------------------------------------------\n")
        
    def step(self):
        self.wealth += self.salary - self.expenses
        print("Nuwan got a salary. His wealth is now ", self.wealth, " after the expenses")
        
#         if random.randint(1,10) > 5:
#              return
        
        
        if not self.bought_a_land:
            if self.model.schedule.agents[1].land_price < self.wealth:
                print("Land price is now ", self.model.schedule.agents[1].land_price)
                print("Nuwan has enough money buy the land. He is going to meet the Land Owner")
                print("Nuwan paid the money and bought the land")
                self.wealth -= self.model.schedule.agents[1].land_price
                self.expenses += self.model.schedule.agents[1].maintain_cost
                self.model.schedule.agents[1].sell_land()
                print("Nuwan now has a land. He spent a lot. Now his wealth is ", self.wealth)
                print("Nuwan's monthly expenses increased to ", self.expenses, " because now he needs to maintain the land")
                print("-------------------------------------------------------\n")
                self.bought_a_land = True
            else:
                print("Nuwan does not have enough money to buy a land yet")
                print("Land price is ",self.model.schedule.agents[1].land_price)
                print("Nuwan only has ",self.wealth, ". He needs to wait")
                print("-------------------------------------------------------\n")
            
            return
                
        if not self.build_a_house:
            if self.model.schedule.agents[2].salary < self.wealth:
                if not self.building_the_house:
                    print("Nuwan found a builder to build his house")
                    print("It takes ", self.model.schedule.agents[2].monthsToBuild, " months to build the house")
                    self.building_the_house = True
                    
                if self.model.schedule.agents[2].monthsToBuild > 0:
                    self.wealth -= self.model.schedule.agents[2].salary
                    self.model.schedule.agents[2].building_the_house()
                    print("After paying the Builder's salary Nuwan now has ", self.wealth)
                    print("-------------------------------------------------------\n")
                    
                else:
                    self.build_a_house = True
                    print("Now Nuwan has a house. But Carpentering is needed.")
                    print("After finishing the masons work now he has ", self.wealth)
                    print("-------------------------------------------------------\n")
                    
            else:
                print("Nuwan does not have enough money to pay the Builder")
                print("-------------------------------------------------------\n")
                
            return
                    
        if not self.house_carpentering:
            if self.model.schedule.agents[3].salary < self.wealth:
                if not self.carpentering_the_house:
                    print("Nuwan found a carpenter to build his house")
                    print("It takes ", self.model.schedule.agents[3].monthsToBuild, " months to do carpentering")
                    self.carpentering_the_house = True
                    
                if self.model.schedule.agents[3].monthsToBuild > 0:
                    self.wealth -= self.model.schedule.agents[3].salary
                    self.model.schedule.agents[3].building_the_house()
                    print("After paying the Carpenter's salary Nuwan now has ", self.wealth)
                    print("-------------------------------------------------------\n")
                    
                else:
                    self.house_carpentering = True
                    print("Now Nuwan has a finished house but needs to paint.")
                    print("After finishing the carpentering now he has ", self.wealth)
                    print("-------------------------------------------------------\n")
                       
            else:
                print("Nuwan does not have enough money to pay the Carpenter")
                print("-------------------------------------------------------\n")
                
            return
            
        if not self.house_painting:
            if self.model.schedule.agents[4].salary < self.wealth:
                if not self.painting_the_house:
                    print("Nuwan found a painter to paint his house")
                    print("It takes ", self.model.schedule.agents[4].monthsToPaint, " months to do painting")
                    self.painting_the_house = True
                    
                if self.model.schedule.agents[4].monthsToPaint > 0:
                    self.wealth -= self.model.schedule.agents[4].salary
                    self.model.schedule.agents[4].painting_the_house()
                    print("After paying the Painter's salary Nuwan now has ", self.wealth)
                    print("-------------------------------------------------------\n")
                    
                else:
                    self.house_painting = True
                    print("Now Nuwan has a fully finished house.")
                    print("After finishing the house now he has ", self.wealth)
                    print("-------------------------------------------------------\n")
                       
            else:
                print("Nuwan does not have enough money to pay the Painter")
                print("-------------------------------------------------------\n")
            
            return
        
        if not self.bought_a_car:
            if self.model.schedule.agents[5].car_price < self.wealth:
                self.wealth -= self.model.schedule.agents[5].car_price
                self.expenses += self.model.schedule.agents[5].maintain_cost
                self.model.schedule.agents[5].sell_car()
                print("Nuwan now has a car. He spent a lot. Now his wealth is ", self.wealth)
                print("Nuwan's monthly expenses increased to ", self.expenses)
                print("!!!!!!!!!!!!!!!!His life is now completed!!!!!!!!!!!!!!!!!!!")
                self.bought_a_car = True
                self.completed = True
            else:
                print("Nuwan does not have enough money to buy a car yet")
                print("Car price is ",self.model.schedule.agents[5].car_price)
                print("Nuwan only has ",self.wealth, ". He needs to wait")
                print("-------------------------------------------------------\n")
            
            return
        
        
class LandOwner(Agent):
    """Land owner. Who is going to sell his land"""
    def __init__(self, unique_id, model):
        super().__init__(unique_id, model)
        self.land_price = random.randint(5000000,7500000)
        self.land_sold = False
        self.maintain_cost = 10000
        self.wealth = 0
        
    def sell_land(self):
        self.wealth = self.land_price
        self.land_price = 0
        self.land_sold = True
        print("Land Owner sold the land. Now the Land Owner has ",self.wealth)
        
    def step(self):
        if not self.land_sold:
            self.land_price += 20000
    
class Builder(Agent):
    """House builder. Who is building houses"""
    def __init__(self, unique_id, model):
        super().__init__(unique_id, model)
        self.salary = random.randint(40000,60000)
        self.monthsToBuild = random.randint(5,10)
        self.wealth = 0
        
    def building_the_house(self):
        self.wealth += self.salary
        self.monthsToBuild -= 1
        if self.monthsToBuild == 0:
            print("Builder finished the house. From building Nuwan's house now he has ", self.wealth)
        else :
            print("Builder did a monthly work. He got a salary. Now builder has ",self.wealth)
            print(self.monthsToBuild," months time work remaining for the builder to finish it.")
                  
    def step(self):
        return
        
class Carpenter(Agent):
    """Carpenter. Who is doing wooden works"""
    def __init__(self, unique_id, model):
        super().__init__(unique_id, model)
        self.salary = random.randint(40000,60000)
        self.monthsToBuild = random.randint(2,5)
        self.wealth = 0
        
    def building_the_house(self):
        self.wealth += self.salary   
        self.monthsToBuild -= 1
        if self.monthsToBuild == 0:
            print("Carpenter finished the his work. From building Nuwan's house now he has ", self.wealth)
        else :
            print("Carpenter did a monthly work. He got a salary. Now carpenter has ",self.wealth)
            print(self.monthsToBuild," months time work remaining for the carpenter to finish it.")
                
    def step(self):
        return
        
class Painter(Agent):
    """Painter. Who is painting houses"""
    def __init__(self, unique_id, model):
        super().__init__(unique_id, model)
        self.salary = random.randint(20000,30000)
        self.monthsToPaint = 1
        self.wealth = 0
        
    def painting_the_house(self):
        self.wealth += self.salary   
        self.monthsToPaint -= 1
        if self.monthsToPaint == 0:
            print("Painter did a monthly work and painted the house. From painting Nuwan's house now he has ", self.wealth)
        else :
            print("Painter did a monthly work. He got a salary. Now painter has ",self.wealth)
            print(self.monthsToPaint," months time work remaining for the painter to finish it.")
        
    def step(self):
        return

class CarSale(Agent):
    """CarSale. Selling Cars"""
    def __init__(self, unique_id, model):
        super().__init__(unique_id, model)
        self.car_price = random.randint(5000000,7500000)
        self.maintain_cost = 10000
        self.wealth = 0
        
    def sell_car(self):
        self.wealth = self.car_price
        self.car_price = 0
        print("CarSale sold the car for ",self.car_price)
        print("Now car sale has a wealth of ", self.wealth)
        
    def step(self):
        return
        
class LifeModel(Model):
    """Life of Nuwan"""
    def __init__(self):
        self.schedule = RandomActivation(self)
        # Create agents
        i = 0
        a = Nuwan(i, self)
        i+=1
        b = LandOwner(i, self)
        i+=1
        c = Builder(i, self)
        i+=1
        d = Carpenter(i, self)
        i+=1
        e = Painter(i, self)
        i+=1
        f = CarSale(i, self)
    
        self.schedule.add(a)
        self.schedule.add(b)
        self.schedule.add(c)
        self.schedule.add(d)
        self.schedule.add(e)
        self.schedule.add(f)
        
    def step(self):
        '''Advance the model by one step.'''
        self.schedule.step()
```
